```python
#zadanie 15


def Grade(df):
  gpa_a = [0]*1000
  gpa_b = [0]*1000
  gpa_c = [0]*1000
  gpa = [0]*1000
  for i in range(1000):
    gpa_a[i] = df.iloc[i, 6]
    gpa_b[i] = df.iloc[i, 7]
    gpa_c[i] = df.iloc[i, 8]
    gpa[i] = (gpa_a[i] + gpa_b[i] + gpa_c[i])/3
    if (gpa[i] < 50):
      gpa[i] = 'F'
    elif (gpa[i] < 60):
      gpa[i] = 'E'
    elif (gpa[i] < 70):
      gpa[i] = 'D'
    elif (gpa[i] < 80):
      gpa[i] = 'C'
    elif (gpa[i] < 90):
      gpa[i] = 'B'
    elif (gpa[i] <= 100):
      gpa[i] = 'A'
  df.insert(9, 'GRADE', gpa, allow_duplicates=True)


Grade(df)
# del df['GRADE']


a = df[(df['GRADE'] == 'A')].count()[0]
b = df[(df['GRADE'] == 'B')].count()[0]
c = df[(df['GRADE'] == 'C')].count()[0]
d = df[(df['GRADE'] == 'D')].count()[0]
e = df[(df['GRADE'] == 'E')].count()[0]
f = df[(df['GRADE'] == 'F')].count()[0]
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
```
